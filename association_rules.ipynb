```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
# conda install -c conda-forge mlxtend
```

```python
book = pd.read_csv("C:\\Users\\sathya\\Documents\\datascience\\book.csv")
```

```python
book.head()
```

```python
df= pd.get_dummies(book)
df.shape
```

```python
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True,max_len=3)
frequent_itemsets.shape
```

```python
confidence_association = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.3)
```

```python
confidence_association.head(10)
```

```python
lift_association = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
lift_association.head(10)
```

```python
redundancy = lift_association.sort_values(by=['leverage','confidence'],axis=0, ascending=False).reset_index()
redundancy = redundancy.drop(['index'], axis=1)
redundancy.shape
redundancy.head()
```

```python
unique_rules = redundancy.iloc[::1]
unique_rules.shape
```

```python
unique_rules.head()
```

```python
top_sort = unique_rules[unique_rules['lift']>1.5]
```

```python
top_sort = top_sort.sort_values(by='lift', ascending=False)
top_sort.reset_index(inplace=True)
top_sort = top_sort.drop(['index'],axis=1)
top_sort.head()
```

```python
#2nd dataset MY_MOVIES
import pandas as pd
groc = pd.read_csv("D:\\brows_dwnlds\\my_movies.csv")
groc.shape
```

```python
groc1 = groc.drop(['V1','V2','V3','V4','V5'],axis=1,inplace =True)
df = pd.get_dummies(groc)
df.head()
```

```python
df.info()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

freq_items = apriori(df, min_support=0.02, use_colnames=True, max_len=3)
```

```python
freq_items.head()
print(freq_items.shape)
```

```python
conf_association = association_rules(freq_items, metric='confidence', min_threshold=0.2)
conf_association.head()
```

```python
lift_association = association_rules(freq_items, metric="lift", min_threshold=1)
print(lift_association.shape)
lift_association.head()
```

```python
# Eliminating redudancy sets
redundancy = lift_association.sort_values(by=['leverage','confidence'],axis=0, ascending=False).reset_index()
redundancy = redundancy.drop(['index'], axis=1)
redundancy.shape
redundancy.head()
```

```python
unique_rules = redundancy.iloc[::2]
unique_rules.shape
```

```python
top_10 = unique_rules[unique_rules['lift']>2]
top_10_sort = top_10.sort_values(by='lift', ascending=False)
top_10_sort.reset_index(inplace=True)
```

```python
top_10_sort = top_10_sort.drop(['index'],axis=1)
```
